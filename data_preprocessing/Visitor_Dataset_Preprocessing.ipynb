Finding similar values from the location dataset for the inconsistant namings

```python
import Levenshtein

def find_closest_matches(list1, list2):
  """Finds the closest match for each item in list1 from list2 based on Levenshtein distance."""
  closest_matches = []
  for item1 in list1:
    distances = [Levenshtein.distance(item1, item2) for item2 in list2]
    closest_index = distances.index(min(distances))
    closest_matches.append(list2[closest_index])
  return closest_matches


list1 = missing_places.tolist()  #inconsistant names
list2 = location_list            #location names from location dataset
list1 = [item for item in list1 if item is not None]  #removing None values

# Find the closest matches
closest_matches = find_closest_matches(list1, list2)
print(closest_matches)
```

```python
# Print lists side by side
print(f"{'List 1':<10}{'List 2':<10}")
for x, y in zip(list1, closest_matches):
    print(f"{x:<10} \t {y:<10}")
```
