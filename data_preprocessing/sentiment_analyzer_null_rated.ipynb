```python
import re
out_list = []
pattern = r'\'\s*,\s*\'|\"\s*,\s*\"'

for review_set in lat_revies:
  temp_list = []
  for review in re.split(pattern, review_set):
    t = re.sub("Ã¢Â€Â™", "'", review.strip("[").strip("]").strip('"').strip("'"))
    temp_list.append(t)
  out_list.append(temp_list)

```

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
print(sentiment_analysis("I love this!"))
```

```python
n = [[sentiment_analysis(t) for t in item] for item in out_list]
```

```python
len(n)
```

```python

positive_scores = []
for item in n:
  temp_score = 0
  for sentiment in item:
    if sentiment[0]['label'] == 'POSITIVE':
      temp_score += sentiment[0]['score']
    elif sentiment[0]['label'] == 'NEGATIVE':
      temp_score += 1 - sentiment[0]['score']
  positive_scores.append(temp_score/len(item))
```

```python
df["emotion_scores"] = positive_scores
```

```python
df = df.drop(subset=['user_rating_total'])
```

```python
df['rating'] = (df['emotion_scores']*5).round(2)
df['user_ratings_total'] = None
```
